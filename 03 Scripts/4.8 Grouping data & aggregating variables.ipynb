```python
# aggregated mean of the “order_number” column grouped by “department_id” 
ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

```python
# creating a new column for max orders of existing customers using the transform() function.

ords_prods_merge['max_order']=ords_prods_merge.groupby('user_id')['order_number'].transform(np.max)
```

```python
# creating a loyalty_flag for existing customers using the max_order column  and loc() functions.
ords_prods_merge.loc[ords_prods_merge['max_order']>40,'Loyalty_Flag']='Loyalty customer'
ords_prods_merge.loc[(ords_prods_merge['max_order']<=40) & (ords_prods_merge['max_order']>10),'Loyalty_Flag']='Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order']<=10,'Loyalty_Flag']='New customer'
```

```python
#to find the frequency of Loyalty_Flag
ords_prods_merge['Loyalty_Flag'].value_counts(dropna=False)
```

```python
#Use the loyalty flag created and check the basic statistics of the product prices for each loyalty category
ords_prods_merge.groupby('Loyalty_Flag').agg({'prices':['mean','max','min','sum']})
```

# Average purchase of Loyalty customer is lower than others.The sum of prices is highest with 'Regular customer', that is due to higher count of Regular customer compare to New and Loyalty customer.

```python
#Create a spending flag for each user based on the average price across all their orders using the following criteria:
#If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

```python
# create column average_price by finding the average of prices per user
ords_prods_merge['average_price']=ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
#create column spending_flag 
ords_prods_merge.loc[ords_prods_merge['average_price']<10,'spending_flag']='Low spender'
ords_prods_merge.loc[ords_prods_merge['average_price']>=10,'spending_flag']='High spender'
```

```python
ords_prods_merge.head(10)
```

```python
#Checking the count of 'spending_flag' values
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
# create column median_freq by finding the average of prices per user
ords_prods_merge['median_freq']=ords_prods_merge.groupby('user_id')['days_since_previous_order'].transform(np.median)
```

```python
#create column order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_freq']>20,'order_frequency_flag']='Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_freq']>10) & (ords_prods_merge['median_freq']<=20) ,'order_frequency_flag']='Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_freq']<=10,'order_frequency_flag']='Frequent customer'
```

```python
#Checking the count of 'order_frequency_flag' values
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
