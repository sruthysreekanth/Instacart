```python
#Drop merge column from df_imported
df_imported=df_imported.drop(columns=['_merge'])
```

```python
# Change types for df_imported data set

df_imported['order_id']=df_imported['order_id'].astype('int32')
df_imported['user_id'] = df_imported['user_id'].astype('int32')
df_imported['order_number']=df_imported['order_number'].astype('int8')
df_imported['order_day_of_week']=df_imported['order_day_of_week'].astype('int8')
df_imported['order_hour_of_day']=df_imported['order_hour_of_day'].astype('int8')
df_imported['days_since_previous_order']=df_imported['days_since_previous_order'].astype('float16')
df_imported['product_id']=df_imported['product_id'].astype('int32')
df_imported['add_to_cart_order']=df_imported['add_to_cart_order'].astype('int32')
df_imported['reordered']=df_imported['reordered'].astype('int8')
```

```python
# Change types for df_prods_clean data set

df_prods_clean['product_id']=df_prods_clean['product_id'].astype('int32')
df_prods_clean['aisle_id'] = df_prods_clean['aisle_id'].astype('int8')
df_prods_clean['department_id']=df_prods_clean['department_id'].astype('int8')
df_prods_clean['prices']=df_prods_clean['prices'].astype('float16')
```

```python
# Merged the dataframes df_imported and df_prods_clean
df_orders_products_merged=df_imported.merge(df_prods_clean,on='product_id',indicator=True)
```
