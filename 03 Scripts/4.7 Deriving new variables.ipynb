# If Statement
# Define a function

```python
def price_label(row):
    if row['prices']<=5:
        return 'Low-range product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Mid-range product'
    elif row['prices']>15:
        return 'High-range product'
    else:
        return 'Not enough data'
```

# Apply the function in dataframe df-sample

```python
df_sample['price_range']=df_sample.apply(price_label,axis=1)
```

# Count the values in new column price_range

```python
df_sample['price_range'].value_counts(dropna=False)
```

# If statement with the loc() function


```python
df_sample.loc[df_sample['prices']<=5,'price_range_loc']='Low-range product'
```

```python
df_sample.loc[(df_sample['prices']>5) & (df_sample['prices']<=15),'price_range_loc']='Medium-range product'
```

```python
df_sample.loc[df_sample['prices']>15,'price_range_loc']='High-range product'
```

# Task 4.7 Deriving New Variables

```python
# Create a new column 'busiest days'
# To find which days are more counts-- 0 is saturday
df_merged['order_day_of_week'].value_counts(dropna=False)

```

```python
# 0 & 1-saturday & sunday are 2 busiest days
# 4 & 3-Wednesday & tuesday are 2 slowest days
# To find the busiest days in instacart

result1=[]
for value1 in df_merged['order_day_of_week']:
    if ((value1==0) or (value1==1)):
        result1.append('Busiest Days')
    elif((value1==4) or (value1==3)):
        result1.append('Slowest days')
    else:
        result1.append('Regular days')
        
```

```python
result1_updated=[]
for value1 in df_merged_updated['order_day_of_week']:
    if ((value1==0) or (value1==1)):
        result1_updated.append('Busiest Days')
    elif((value1==4) or (value1==3)):
        result1_updated.append('Slowest days')
    else:
        result1_updated.append('Regular days')
```

```python
# create a new column busiest_days and set it equal to result1
df_merged['busiest_days']=result1
```

```python
#  To find the busiest hours of the day. 

df_merged['order_hour_of_day'].value_counts(dropna=False)
        
```

   # Observations: The day can be split into three parts. 'Most orders' (hour- 10, 11, 14, 15, 13, 12, 16, 9 ie from 9am to 4pm), 'Average orders' (hour- 17, 8, 18, 19, 20, 7, 21, 22 - from 7am to 8am and from 5pm to 10pm), or 'Fewest orders' (hour 23, 6, 0, 1, 5, 2, 4, 3 - from 11pm to 6am).

```python
# find the busiest hours of the day
hours=[]
for hour in df_merged['order_hour_of_day']:
    if (hour in [10, 11, 14, 15, 13, 12, 16, 9]):
        hours.append('Most Orders')
    elif(hour in [23, 6, 0, 1, 5, 2, 4, 3 ]):
        hours.append('Fewest Orders')
    else:
        hours.append('Average Orders')
```

```python
hours_updated=[]
for hour in df_merged_updated['order_hour_of_day']:
    if (hour in [10, 11, 14, 15, 13, 12, 16, 9]):
        hours_updated.append('Most Orders')
    elif(hour in [23, 6, 0, 1, 5, 2, 4, 3 ]):
        hours_updated.append('Fewest Orders')
    else:
        hours_updated.append('Average Orders')
```

```python
hours
```

```python
# create new column busiest_period_of_day.
df_merged['busiest_period_of_day']=hours
```

```python
df_merged_updated['busiest_period_of_day']=hours_updated
```
